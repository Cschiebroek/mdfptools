```python
import pandas as pd
import matplotlib.pyplot as plt
import glob

def plot_density_vs_timestep(confid_start, confid_end):

    for confid in range(confid_start, confid_end + 1):

        filename = f'prod_dat_euler/equilibration_{confid}.dat'
        
        # Read the data
        try:
            df = pd.read_csv(filename)
            plt.figure(figsize=(10, 6))
            plt.plot(df['#"Step"'], df['Density (g/mL)'], marker='o', linestyle='-', markersize=2)
            plt.xlabel('Timestep')
            plt.ylabel('Density (g/mL)')
            plt.title(str(confid))
            #set limit y axis
            plt.ylim(0.5, 3.0)
            plt.show()

        except FileNotFoundError:
            continue
```

```python
plot_density_vs_timestep(17265, 20000)
```
