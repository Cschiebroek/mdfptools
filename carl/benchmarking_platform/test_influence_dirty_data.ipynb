```python
import psycopg2 
import logging
from utils.data_preprocessing import prepare_data
conn = psycopg2.connect("dbname=cs_mdfps user=cschiebroek host=lebanon")
logging.info("Preparing data...")
df = prepare_data(conn,'RDKit_PhysChem')
df
```

```python
df_vp_mid = df[(df['vp_log10_pa'] > 1) & (df['vp_log10_pa'] < 8)]
from sklearn.model_selection import RepeatedKFold
from models.neural_network import NeuralNetworkModel
from utils.data_preprocessing import get_features
ys,preds_all,preds_mid,inchikeys_test = [],[],[],[]
#do a 5x3 cross validation
kf = RepeatedKFold(n_splits=5, n_repeats=3, random_state=2841)
for train_index, test_index in kf.split(df_vp_mid):
    print('Split:',len(ys)+1)
    test_inchikeys = df_vp_mid.iloc[test_index]['molregno'].tolist()
    inchikeys_test.append(test_inchikeys)
    train_all_data = df[~df['molregno'].isin(test_inchikeys)]
    train_mid_data = df_vp_mid[~df_vp_mid['molregno'].isin(test_inchikeys)] 
    test_df = df_vp_mid.iloc[test_index]

    train_X_all, val_X_all = get_features(train_all_data, test_df, descriptor_name='RDKit_PhysChem',scale=True)
    train_y_all = train_all_data['vp_log10_pa']
    val_y = test_df['vp_log10_pa']

    train_X_mid, val_X_mid = get_features(train_mid_data, test_df, descriptor_name='RDKit_PhysChem',scale=True)
    train_y_mid = train_mid_data['vp_log10_pa']
    
    model_all = NeuralNetworkModel(input_shape=(train_X_all.shape[1],))
    model_mid = NeuralNetworkModel(input_shape=(train_X_mid.shape[1],))

    model_all.train(train_X_all, train_y_all, validation_data=(val_X_all, val_y))
    model_mid.train(train_X_mid, train_y_mid, validation_data=(val_X_mid, val_y))
    y_pred_model_all = model_all.predict(val_X_all)
    y_pred_model_mid = model_mid.predict(val_X_mid)
    ys.append(val_y)
    preds_all.append(y_pred_model_all)
    preds_mid.append(y_pred_model_mid)
```

```python
from utils.visualization import density_plots
density_plots(reals_list=[ys,ys],predictions_list=[preds_all,preds_mid],title=['Trained on full range','Trained on mid range'],molregnos_list=[inchikeys_test,inchikeys_test],dims=(1,2))
```

```python
#there seems to be one huge outlier for the mid range one, lets remove it 
ys_no_outier,preds_all_no_outier,preds_mid_no_outier,inchikeys_test_no_outlier = [],[],[],[]
for y,pred_mid,pred_all,molregno in zip(ys,preds_mid,preds_all,inchikeys_test):
    mask = pred_mid < 10
    y = y[mask]
    pred_mid = pred_mid[mask]
    pred_all = pred_all[mask]
    molregno = [molregno[i] for i in range(len(molregno)) if mask[i]]

    ys_no_outier.append(y)
    preds_mid_no_outier.append(pred_mid)
    inchikeys_test_no_outlier.append(molregno)
    preds_all_no_outier.append(pred_all)
density_plots(reals_list=[ys_no_outier,ys_no_outier],predictions_list=[preds_all_no_outier,preds_mid_no_outier],title=['Trained on full range','Trained on mid range'],molregnos_list=[inchikeys_test_no_outlier,inchikeys_test_no_outlier],dims=(1,2))
```

```python
inchikeys_test_rr,ys_rr,preds_all_rr,preds_mid_rr = [],[],[],[]
from sklearn.linear_model import LinearRegression,Ridge
for train_index, test_index in kf.split(df_vp_mid):
    test_inchikeys = df_vp_mid.iloc[test_index]['molregno'].tolist()
    inchikeys_test_rr.append(test_inchikeys)
    train_all_data = df[~df['molregno'].isin(test_inchikeys)]
    train_mid_data = df_vp_mid[~df_vp_mid['molregno'].isin(test_inchikeys)] 
    test_df = df_vp_mid.iloc[test_index]

    train_X_all, val_X_all = get_features(train_all_data, test_df, descriptor_name='RDKit_PhysChem',scale=True)
    train_y_all = train_all_data['vp_log10_pa']
    val_y = test_df['vp_log10_pa']

    train_X_mid, val_X_mid = get_features(train_mid_data, test_df, descriptor_name='RDKit_PhysChem',scale=True)
    train_y_mid = train_mid_data['vp_log10_pa']
    
    model_all = Ridge(alpha=1)
    model_mid = Ridge(alpha=1)
    model_all.fit(train_X_all, train_y_all)
    model_mid.fit(train_X_mid, train_y_mid)
    y_pred_model_all = model_all.predict(val_X_all)
    y_pred_model_mid = model_mid.predict(val_X_mid)
    ys_rr.append(val_y)
    preds_all_rr.append(y_pred_model_all)
    preds_mid_rr.append(y_pred_model_mid)
```

```python
density_plots(reals_list=[ys_rr,ys_rr],predictions_list=[preds_all_rr,preds_mid_rr],title=['Trained on full range','Trained on mid range'],molregnos_list=[inchikeys_test_rr,inchikeys_test_rr],dims=(1,2),bounds=(-1,10))
```

```python
inchikeys_test_lr,ys_lr,preds_all_lr,preds_mid_lr = [],[],[],[]
from sklearn.linear_model import LinearRegression,Ridge
for train_index, test_index in kf.split(df_vp_mid):
    test_inchikeys = df_vp_mid.iloc[test_index]['molregno'].tolist()
    inchikeys_test_lr.append(test_inchikeys)
    train_all_data = df[~df['molregno'].isin(test_inchikeys)]
    train_mid_data = df_vp_mid[~df_vp_mid['molregno'].isin(test_inchikeys)] 
    test_df = df_vp_mid.iloc[test_index]

    train_X_all, val_X_all = get_features(train_all_data, test_df, descriptor_name='RDKit_PhysChem',scale=True)
    train_y_all = train_all_data['vp_log10_pa']
    val_y = test_df['vp_log10_pa']

    train_X_mid, val_X_mid = get_features(train_mid_data, test_df, descriptor_name='RDKit_PhysChem',scale=True)
    train_y_mid = train_mid_data['vp_log10_pa']
    
    model_all = Ridge(alpha=1)
    model_mid = Ridge(alpha=1)
    model_all.fit(train_X_all, train_y_all)
    model_mid.fit(train_X_mid, train_y_mid)
    y_pred_model_all = model_all.predict(val_X_all)
    y_pred_model_mid = model_mid.predict(val_X_mid)
    ys_lr.append(val_y)
    preds_all_lr.append(y_pred_model_all)
    preds_mid_lr.append(y_pred_model_mid)
```

```python
density_plots(reals_list=[ys_lr,ys_lr],predictions_list=[preds_all_lr,preds_mid_lr],title=['Trained on full range','Trained on mid range'],molregnos_list=[inchikeys_test_lr,inchikeys_test_lr],dims=(1,2),bounds=[-2,10])
```

```python
import numpy as np
from scipy.stats import kendalltau

# Neural Network Model
ktaus_mid_nn, ktaus_all_nn = {}, {}
for y, pred_all, pred_mid, molregno in zip(ys, preds_all, preds_mid, inchikeys_test):
    for i in range(1, 5):
        lower_bound = i
        upper_bound = i + 3
        mask = (y > lower_bound) & (y < upper_bound)
        y_range = y[mask]
        pred_all_range = pred_all[mask]
        pred_mid_range = pred_mid[mask]
        ktau_all = kendalltau(y_range, pred_all_range)[0]
        ktau_mid = kendalltau(y_range, pred_mid_range)[0]
        ktaus_mid_nn[(lower_bound, upper_bound)] = ktaus_mid_nn.get((lower_bound, upper_bound), []) + [ktau_mid]
        ktaus_all_nn[(lower_bound, upper_bound)] = ktaus_all_nn.get((lower_bound, upper_bound), []) + [ktau_all]

mean_ktaus_all_nn, mean_ktaus_mid_nn = {}, {}
for key in ktaus_all_nn.keys():
    mean_ktaus_all_nn[key] = np.mean(ktaus_all_nn[key])
    mean_ktaus_mid_nn[key] = np.mean(ktaus_mid_nn[key])

# Linear Regression Model
ktaus_mid_lr, ktaus_all_lr = {}, {}
for y, pred_all, pred_mid, molregno in zip(ys_lr, preds_all_lr, preds_mid_lr, inchikeys_test_lr):
    for i in range(1, 5):
        lower_bound = i
        upper_bound = i + 3
        mask = (y > lower_bound) & (y < upper_bound)
        y_range = y[mask]
        pred_all_range = pred_all[mask]
        pred_mid_range = pred_mid[mask]
        ktau_all = kendalltau(y_range, pred_all_range)[0]
        ktau_mid = kendalltau(y_range, pred_mid_range)[0]
        ktaus_mid_lr[(lower_bound, upper_bound)] = ktaus_mid_lr.get((lower_bound, upper_bound), []) + [ktau_mid]
        ktaus_all_lr[(lower_bound, upper_bound)] = ktaus_all_lr.get((lower_bound, upper_bound), []) + [ktau_all]

mean_ktaus_all_lr, mean_ktaus_mid_lr = {}, {}
for key in ktaus_all_lr.keys():
    mean_ktaus_all_lr[key] = np.mean(ktaus_all_lr[key])
    mean_ktaus_mid_lr[key] = np.mean(ktaus_mid_lr[key])

mean_ktaus_all_nn, mean_ktaus_mid_nn, mean_ktaus_all_lr, mean_ktaus_mid_lr
```

```python
#plot 2 subplots; the two nns, for each range the mean ktau and 90% confidence interval, and the same for the lr models
import matplotlib.pyplot as plt
import seaborn as sns
fig,ax = plt.subplots(2,1,figsize=(10,10))
#side to side bars

# Neural Network
keys = list(mean_ktaus_all_nn.keys())
keys.sort()
mean_all_nn = [mean_ktaus_all_nn[key] for key in keys]
mean_mid_nn = [mean_ktaus_mid_nn[key] for key in keys]
upper_ci_all_nn = [np.percentile(ktaus_all_nn[key], 95) for key in keys]
lower_ci_all_nn = [np.percentile(ktaus_all_nn[key], 5) for key in keys]
upper_ci_mid_nn = [np.percentile(ktaus_mid_nn[key], 95) for key in keys]
lower_ci_mid_nn = [np.percentile(ktaus_mid_nn[key], 5) for key in keys]

# Linear Regression
mean_all_lr = [mean_ktaus_all_lr[key] for key in keys]
mean_mid_lr = [mean_ktaus_mid_lr[key] for key in keys]
upper_ci_all_lr = [np.percentile(ktaus_all_lr[key], 95) for key in keys]
lower_ci_all_lr = [np.percentile(ktaus_all_lr[key], 5) for key in keys]
upper_ci_mid_lr = [np.percentile(ktaus_mid_lr[key], 95) for key in keys]
lower_ci_mid_lr = [np.percentile(ktaus_mid_lr[key], 5) for key in keys]

x = np.arange(len(keys))

ax[0].bar(x-0.2, mean_all_nn, yerr=[np.array(mean_all_nn) - np.array(lower_ci_all_nn), np.array(upper_ci_all_nn) - np.array(mean_all_nn)], capsize=5, width=0.4, label='All range NN', color='blue')
ax[0].bar(x+0.2, mean_mid_nn, yerr=[np.array(mean_mid_nn) - np.array(lower_ci_mid_nn), np.array(upper_ci_mid_nn) - np.array(mean_mid_nn)], capsize=5, width=0.4, label='Mid range NN', color='orange')

ax[0].set_xticks(x)
ax[0].set_xticklabels([str(key) for key in keys])
ax[0].set_title('Neural Network')
ax[0].set_ylabel('Kendall Tau')
ax[0].legend()

ax[1].bar(x-0.2, mean_all_lr, yerr=[np.array(mean_all_lr) - np.array(lower_ci_all_lr), np.array(upper_ci_all_lr) - np.array(mean_all_lr)], capsize=5, width=0.4, label='All range LR', color='blue')
ax[1].bar(x+0.2, mean_mid_lr, yerr=[np.array(mean_mid_lr) - np.array(lower_ci_mid_lr), np.array(upper_ci_mid_lr) - np.array(mean_mid_lr)], capsize=5, width=0.4, label='Mid range LR', color='orange')

ax[1].set_xticks(x)
ax[1].set_xticklabels([str(key) for key in keys])

ax[1].set_title('Linear Regression')
ax[1].set_ylabel('Kendall Tau')
ax[1].legend()
plt.tight_layout()
plt.show()
```
