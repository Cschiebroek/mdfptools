```python
import os
import numpy as np
from rdkit import Chem
from rdkit.Chem import rdMolDescriptors
from sklearn.linear_model import LinearRegression

# Use the provided atom contribution calculation functions
def calc_atom_contributions(mol):
    """
    Calculate atom-wise LogP and MR contributions for a given molecule.
    """
    # Calculate contributions using RDKit's CalcCrippenDescriptors
    contribs = rdMolDescriptors._CalcCrippenContribs(mol)
    return contribs

def calc_total_properties(mol):
    """
    Calculate total LogP and MR for a molecule.
    """
    contribs = calc_atom_contributions(mol)
    logp, mr = np.sum(contribs, axis=0)  # Sum contributions across all atoms
    return logp, mr

def fit_linear_regression(smiles_list, target_logp):
    """
    Fit a linear regression model based on atom contributions and target values.
    """
    X = []
    for smi in smiles_list:
        mol = Chem.MolFromSmiles(smi)
        atom_contribs = calc_atom_contributions(mol)
        flattened_contribs = np.ravel(atom_contribs)  # Flatten to a 1D feature array
        X.append(flattened_contribs)
    
    X = np.array(X)  # Convert to NumPy array for fitting
    y = np.array(target_logp)  # Target LogP values

    # Fit linear regression
    model = LinearRegression()
    model.fit(X, y)

    return model

# Example usage
if __name__ == '__main__':
    # Sample molecules (SMILES) and their experimental LogP values
    smiles_list = ['CCO', 'CC', 'CCCC', 'C1CCCCC1']  # Add more SMILES as needed
    target_logp = [0.34, 0.27, 1.95, 2.47]  # Add corresponding experimental LogP values

    # Fit linear regression model
    model = fit_linear_regression(smiles_list, target_logp)
    
    print("Coefficients:", model.coef_)
    print("Intercept:", model.intercept_)
```
