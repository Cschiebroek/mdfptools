```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
df_ochem = pd.read_csv('/localhome/cschiebroek/ShuZe/vp/data cleaning/ochem_vp.tsv', sep='\t')
print(len(df_ochem))
#convert all temperatures to Kelvin
```

```python
#change every = x to x for Temperature
df_ochem['Temperature'] = df_ochem['Temperature'].str.replace('=', '')
#make floats
df_ochem['Temperature'] = df_ochem['Temperature'].astype(float)
```

```python
df_ochem['Temperature_K'] = df_ochem.apply(lambda row: row['Temperature'] + 273.15 if 'C' in row[df_ochem.columns[5]] else row['Temperature'], axis=1)
df_ochem
```

```python
#rename column 2 to VP_measured and 3 to VP_unit
df_ochem = df_ochem.rename(columns={df_ochem.columns[2]: 'VP_measured', df_ochem.columns[3]: 'VP_unit'})
```

```python
df_ochem[df_ochem.columns[2]] = df_ochem[df_ochem.columns[2]].str.replace('<', '')
```

```python
df_ochem
```

```python
#convert all values in second column to floats
df_ochem[df_ochem.columns[2]] = df_ochem[df_ochem.columns[2]].astype(float)
```

```python
# # where unit is log(Pa), convert to Pa
# df_ochem['VP_pa'] = df_ochem.apply(lambda row: 10**row[df_ochem.columns[2]] if row[df_ochem.columns[3]] == 'log(Pa)' else row[df_ochem.columns[2]], axis=1)
# df_ochem['VP_pa'] = df_ochem.apply(lambda row: row[df_ochem.columns[2]]*133.322 if row[df_ochem.columns[3]] == 'mmHg' else row[df_ochem.columns[2]], axis=1)
# df_ochem['VP_pa'] = df_ochem.apply(lambda row: 133.322*10**row[df_ochem.columns[2]] if row[df_ochem.columns[3]] == 'log(mmHg)' else row[df_ochem.columns[2]], axis=1)
# df_ochem['VP_pa'] = df_ochem.apply(lambda row: row[df_ochem.columns[2]]/1000 if row[df_ochem.columns[3]] == 'mPa' else row[df_ochem.columns[2]], axis=1)
# df_ochem['VP_pa'] = df_ochem.apply(lambda row: row[df_ochem.columns[2]]*1000 if row[df_ochem.columns[3]] == 'kPa' else row[df_ochem.columns[2]], axis=1)

unit_conversion = {
    'hPa': 100,
    'mPa' : 0.001,
    'Pa': 1,
    'mm Hg': 133.322,
    'mmHg' : 133.322,
    'kPa': 1000,
    'mBar': 100,
    'Torr': 133.322,
    'Bar': 100000,
    'PSI': 6894.76,
    'atm': 101325,
    'log10(mmHg)': pd.NA,
    'E-03 Pa': 0.001,
    'mg/mÎ_ air': 0.000001,
    'µPa': 1e-6,
    'E-10 Pa': 1e-10,
    'E-05 Pa': 1e-5,
    'Nm-2': 1,
    'e-20 Pa': 1e-20,
    'x10-4 mmHg': 0.000133322,
    'MPa': 1e6,
    'E-8 Pa': 1e-8,
    'E-7 Pa': 1e-7,
    'no data': pd.NA,
    'E -6 Pa': 1e-6,
    'E-6 Pa': 1e-6,
    '10-3 Pa': 0.001,
    'see below': pd.NA,
    #log(Pa), keep as is
    'log(Pa)': pd.NA,
    'ln(Pa)' : pd.NA,
    'log(mmHg)': pd.NA,
}
df_non_log = df_ochem[df_ochem[df_ochem.columns[3]].str.contains('log') == False]

df_non_log['VP_Pa'] = df_non_log.apply(lambda row: row[df_non_log.columns[2]] * unit_conversion[row[df_non_log.columns[3]]], axis=1)
df_non_log

df_log = df_ochem[df_ochem[df_ochem.columns[3]].str.contains('log') == True]
df_log[df_ochem.columns[3]].value_counts()
#make two new rows, one with the non log value, one with the unit.
df_log['VP_val_nonlog'] = df_log[df_log.columns[2]]
df_log['VP_val_nonlog'] = df_log.apply(lambda row: 10**row['VP_val_nonlog'], axis=1)
df_log['VP_unit_nonlog'] = df_log[df_log.columns[3]]
#replace('log(Pa)', 'Pa')
df_log['VP_unit_nonlog'] = df_log.apply(lambda row: row['VP_unit_nonlog'].replace('log', ''), axis=1)
df_log['VP_unit_nonlog'] = df_log.apply(lambda row: row['VP_unit_nonlog'].replace('(', ''), axis=1)
df_log['VP_unit_nonlog'] = df_log.apply(lambda row: row['VP_unit_nonlog'].replace(')', ''), axis=1)
df_log['VP_unit_nonlog'] = df_log.apply(lambda row: row['VP_unit_nonlog'].replace(' ', ''), axis=1)
#do unit conversion
df_log['VP_Pa'] = df_log.apply(lambda row: row['VP_val_nonlog'] * unit_conversion[row['VP_unit_nonlog']], axis=1)
#show nan values
df_log[df_log['VP_Pa'].isna()]
df_log
#drop columns 2 and 3
df_log = df_log.drop(df_log.columns[[2, 3]], axis=1)
#rename columns
df_log = df_log.rename(columns={'VP_val_nonlog': 'VP_measured', 'VP_unit_nonlog': 'VP_unit'})
df_log
```

```python
#combine them
df_combined = pd.concat([df_non_log, df_log])
df_combined
df_ochem = df_combined
#check where VP_Pa is nan
df_ochem[df_ochem['VP_Pa'].isna()]['VP_unit'].value_counts()    
```

```python
#make smiles canonical and standardized
from rdkit import Chem
from rdkit.Chem.MolStandardize import rdMolStandardize
df_ochem['standardized_smiles'] = df_ochem.apply(lambda row: Chem.MolToSmiles(rdMolStandardize.ChargeParent(Chem.MolFromSmiles((row[df_ochem.columns[1]])))), axis=1)
```

```python
from rdkit import Chem
df_ochem['NumberOfHeavyAtoms'] = df_ochem.apply(lambda row: Chem.MolFromSmiles(row['standardized_smiles']).GetNumHeavyAtoms(), axis=1)
#plot NumberOfHeavyAtoms against VP_pa
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(data=df_ochem, x='NumberOfHeavyAtoms', y='VP_pa')
#yscale('log')
plt.yscale('log')
plt.show()
```

```python
#new column with log10(VP_pa)
import numpy as np
df_ochem['log10VP_pa'] = df_ochem.apply(lambda row: np.log10(row['VP_pa']), axis=1)
```

```python
df_ochem.to_csv('OCHEM_standardized_not_cleaned.csv', index=False)
```
