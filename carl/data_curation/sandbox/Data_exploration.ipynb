```python
import pandas as pd
```

```python
df_original = pd.read_csv('/localhome/cschiebroek/ShuZe/vp/data cleaning/cleaned_vp_all.tsv', sep='\t')
#This dataframe can still be found on the harddrive, as well as on Euler in Shuze's work directory
print("Original amount of datapoints: ", len(df_original))
print(df_original['Temperature'].value_counts())   
#As we see. this dataframe has 4436 datapoints, most of which (3098) have a temperature of 298.15K
df_original_298 = df_original[df_original['Temperature'] == 298.15] 
```

```python
from rdkit import Chem
from rdkit.Chem.MolStandardize import rdMolStandardize
df_original_298['standardized_smiles'] = df_original_298.apply(lambda row: Chem.MolToSmiles(rdMolStandardize.ChargeParent(Chem.MolFromSmiles((row['SMILES'])))), axis=1)
```

```python
df_original_298['inchi'] = df_original_298.apply(lambda row: Chem.MolToInchi(Chem.MolFromSmiles(row['standardized_smiles'])), axis=1)
df_original_298['inchikey'] = df_original_298.apply(lambda row: Chem.MolToInchiKey(Chem.MolFromSmiles(row['standardized_smiles'])), axis=1)
```

```python
df_opera_25 = pd.read_csv('../opera_full_25.csv')
df_opera_25['inchi'] = df_opera_25.apply(lambda row: Chem.MolToInchi(Chem.MolFromSmiles(row['SMILES'])), axis=1)
df_opera_25['inchikey'] = df_opera_25.apply(lambda row: Chem.MolToInchiKey(Chem.MolFromSmiles(row['SMILES'])), axis=1)
df_opera_25
```

```python
#check for different matching how many matches: standardized_smiles, inchi, inchikey
print(df_opera_25['inchi'].isin(df_original_298['inchi']).value_counts())
print(df_opera_25['inchikey'].isin(df_original_298['inchikey']).value_counts())
print(df_opera_25['SMILES'].isin(df_original_298['standardized_smiles']).value_counts())
```

```python
#make three sets: 1. overlap 2. opera 3. shuze
overlap = df_opera_25[df_opera_25['inchi'].isin(df_original_298['inchi'])]
opera = df_opera_25[~df_opera_25['inchi'].isin(df_original_298['inchi'])]
shuze = df_original_298[~df_original_298['inchi'].isin(df_opera_25['inchi'])]
#print descriptino and legnht
print("overlap: ", len(overlap))
print("opera: ", len(opera))
print("shuze: ", len(shuze))
all_inchies = pd.concat([overlap, opera, shuze])
print("all: ", len(all_inchies))
```
