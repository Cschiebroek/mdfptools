```python
import pandas as pd

def convert_log10_mmHg(x):
    return 10 ** float(x) * 133.322  # Convert log10(mmHg) to Pa

unit_conversion = {
    'hPa': 100,
    'Pa': 1,
    'mm Hg': 133.322,
    'kPa': 1000,
    'mBar': 100,
    'Torr': 133.322,
    'Bar': 100000,
    'PSI': 6894.76,
    'atm': 101325,
    'log10(mmHg)': pd.NA,
    'E-03 Pa': 0.001,
    'mg/mÎ_ air': 0.000001,
    'µPa': 1e-6,
    'E-10 Pa': 1e-10,
    'E-05 Pa': 1e-5,
    'Nm-2': 1,
    'e-20 Pa': 1e-20,
    'x10-4 mmHg': 0.000133322,
    'MPa': 1e6,
    'E-8 Pa': 1e-8,
    'E-7 Pa': 1e-7,
    'no data': pd.NA,
    'E -6 Pa': 1e-6,
    'E-6 Pa': 1e-6,
    '10-3 Pa': 0.001,
    'see below': pd.NA
}

df_reach['VP_Pa'] = df_reach.apply(lambda row: row['P (lowerValue)'] * unit_conversion.get(row['P (unit)'], 1), axis=1)
#drop Na
df_reach = df_reach.dropna(subset=['VP_Pa'])
df_reach
```

```python
#make lower value float
df_reach['T (lowerValue)'] = df_reach['T (lowerValue)'].astype(float)
```

```python
df_reach['Temperature_K'] = df_reach.apply(lambda row: row['T (lowerValue)'] + 273.15 if isinstance(row['T (unit)'], str) and 'C' in row['T (unit)'] else row['T (lowerValue)'], axis=1)
```

```python
# result_df['cas'].value_counts()
import pickle
print(len(df_reach))
cas2smiles = pickle.load(open("/localhome/cschiebroek/ShuZe/vp/data cleaning/cas_id_2_smiles.pickle", "rb"))
cas2smiles_dict = {i:j for i,j in cas2smiles}
df_reach['SMILES'] = df_reach['cas'].map(cas2smiles_dict)
#make smiles canonical and standardized
from rdkit import Chem
from rdkit.Chem.MolStandardize import rdMolStandardize
#drop where smiles = None
df_reach = df_reach[[i != None for i in df_reach.SMILES]]
#and no Nan in smiles
df_reach = df_reach.dropna(subset=['SMILES'])
smiles = df_reach['SMILES'].tolist()
mols = [Chem.MolFromSmiles(i) for i in smiles]
df_reach = df_reach[[i != None for i in mols]]
df_reach['standardized_smiles'] = df_reach.apply(lambda row: Chem.MolToSmiles(rdMolStandardize.ChargeParent(Chem.MolFromSmiles((row['SMILES'])))), axis=1)
print(len(df_reach))
```

```python
#change VP_Pa to log10(VP_Pa)
import numpy as np
df_reach['log10_VP_Pa'] = df_reach.apply(lambda row: np.log10(row['VP_Pa']), axis=1)
```

```python
df_reach['NumHeavyAtoms'] = df_reach.apply(lambda row: Chem.MolFromSmiles(row['standardized_smiles']).GetNumHeavyAtoms(), axis=1)
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(data=df_reach, x='NumHeavyAtoms', y='log10_VP_Pa')
#yscale('log')
plt.show()
#yscale('log')
plt.show()
```
