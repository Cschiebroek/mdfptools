### Session Creation
- Limit rate of creation requests
- Any application shall not exceed 5 concurrent session requests per second
- Limit number of concurrent active sessions
- Any application shall not exceed 10 concurrent active sessions
### Search and Retrieve
- Limit rate of search requests
- Any application shall not exceed 100 search requests per second per session
- Limit rate of retrieve requests
- Any application shall not exceed 100 retrieve requests per second per session
### Resultsets
- Limit number of resultsets
- Any application shall not exceed 500 resultsets per session

```python
def retrieve_vp_from_cas(ra,search_and_retrieve_requests,number_of_resultssets,cas_ids):
    for cas_id in cas_ids:
        ra.select("RX", "S", "IDE.RN = '" + cas_id + "'", "",
                  "WORKER,NO_CORESULT")
        if ra.resultsize != 0:
            response = ra.retrieve(ra.resultname, ["VP"], str(1),
                                   str(ra.resultsize), "", "", "", "")
            results = ra.get_field_content(response, "VP.VP")
            for r in results:
                # datapoint = cas_id + "\t" + r + "\n"
                print(f'CAS: {cas_id} has vapor pressur of {r} Torr')
    number_of_resultssets += 1
    search_and_retrieve_requests += 1
    print("Number of results sets: ", number_of_resultssets)
    print("Search and retrieve requests: ", search_and_retrieve_requests)
    return search_and_retrieve_requests, number_of_resultssets,results

def get_vp_and_temp_from_cas(ra, search_and_retrieve_requests, number_of_resultssets, cas_ids):
    results = []
    for cas_id in cas_ids:
        ra.select("RX", "S", "IDE.RN = '" + cas_id + "'", "", "WORKER,NO_CORESULT")
        if ra.resultsize != 0:
            response = ra.retrieve(ra.resultname, ["VP"], str(1), str(ra.resultsize), "", "", "", "")
            vp_results = ra.get_field_content(response, "VP.VP")
            vp_tmp = ra.get_field_content(response, "VP.T")
            for vp, t in zip(vp_results, vp_tmp):
                results.append((cas_id, vp, t))
                print(f"CAS: {cas_id} has vapor pressure of {vp} Torr at temperature {t} K")
    number_of_resultssets += 1
    search_and_retrieve_requests += 1
    print("Number of result sets: ", number_of_resultssets)
    print("Search and retrieve requests: ", search_and_retrieve_requests)
    return search_and_retrieve_requests, number_of_resultssets, results
```
